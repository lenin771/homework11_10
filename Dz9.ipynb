# 1-дневное удержание по AB-группе

```python
df.groupby('version')['retention_1'].mean()
```

*Через сутки после установки игры возвращается чуть меньше половины игроков. Посмотрим, как отличается однодневное удержание между двумя AB-группами*

*Наблюдаем небольшое снижение однодневного удержания, при перемещении ворот на уровень 40.*

```python
boot_1d = []
iterations=500
for i in range(iterations):
    boot_mean = df.sample(frac=1, replace=True).groupby(
        'version')['retention_1'].mean()
    boot_1d.append(boot_mean)
    
boot_1d = pd.DataFrame(boot_1d)
    
ax = boot_1d.plot.kde()
ax.set_title('Начальные 1-дневные удержание для каждой AB-группы')
ax.set_xlabel('Среднее удержание за 1 день');
```

*Из графика видно, что есть некоторые признаки разницы.*

```python
boot_1d.head()
```

Добавим столбец с разницей в % между двумя AB-группами

```python
boot_1d['diff'] = ((boot_1d.gate_30 - boot_1d.gate_40)
                   / boot_1d.gate_40 * 100)
```

График разницы в % начальной загрузки

```python
ax = boot_1d['diff'].plot.kde()
ax.set_xlabel('Средняя разница')
ax.set_title('% Разница в 1-дневном удержании');
```

# Вероятность разницы

*На этой диаграмме мы видим, что наиболее вероятная разница в % составляет около 1% - 2%, и что большая часть распределения выше 0% в пользу ворот на уровне 30.*

Расчет вероятности того, что 1-дневное удержание выше, когда ворота находятся на уровне 30

```python
prob = (boot_1d['diff'] > 0).sum() / len(boot_1d['diff'])
```

```python
print(f'{round(prob * 100, 2)} %')
```

# 7-дневное удержание по AB-группе

*Видно, что 1-дневное удержание лучше, когда ворота находятся на уровне 30. Однако, поскольку игроки играют в игру только один день, вполне вероятно, что большинство игроков еще не достигли 30 уровня. То есть многие игроки не будут затронуты вратами, даже если они уже на 30-м уровне.*

Расчитаем 7-дневного удержания для двух групп AB

```python
df.groupby('version')['retention_7'].mean()
```

*Как и в случае с 1-дневным удержанием, мы видим, что 7-дневное удержание немного ниже, когда ворота находятся на уровне 40, чем когда ворота находятся на уровне 30. Эта разница также больше, чем для 1-дневного удержания, по-видимому, потому, что больше игроков успели попасть в первые ворота.*

Создаем список и преобразуем списка в DataFrame.

```python
boot_7d = []
for i in range(500):
    boot_mean = df.sample(frac=1, replace=True).groupby(
        'version')['retention_7'].mean()
    boot_7d.append(boot_mean)
    
boot_7d = pd.DataFrame(boot_7d)
```

Добавление столбца с разницей в % между двумя AB-группами

```python
boot_7d['diff'] = ((boot_7d['gate_30'] - boot_7d['gate_40']) / 
                   boot_7d['gate_40'] * 100)
```

Построения графика

```python
ax = boot_7d['diff'].plot.kde()
ax.set_xlabel("% difference in means")
ax.set_title('Bootstrap % Difference in 7-Day Retention Means');
```

Расчет вероятности, что 7-дневное удержание выше на уровне 30

```python
prob = (boot_7d['diff'] > 0).mean()
print(f'{round(prob * 100, 2)} %')
```

Результат говорит нам о том, что существуют убедительные доказательства того, что 7-дневное удержание выше, когда ворота находятся на уровне 30, чем когда они находятся на уровне 40.
